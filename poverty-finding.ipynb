```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')

df = pd.read_excel('https://github.com/tiecia/inequality-project/raw/refs/heads/main/DistrictCostDatabase_2024.xlsx', sheet_name='Data')
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

poverty = df["pov"].values.reshape(-1, 1)
fundinggap = df["fundinggap"].values

# Perform linear regression
model = LinearRegression().fit(poverty, fundinggap)

# Predict values for the line of best fit
y_pred = model.predict(poverty)

plt.scatter(poverty, fundinggap)
plt.plot(poverty, y_pred, color='red')
plt.xlabel('Child Poverty Rate')
plt.ylabel('Average Funding Gap Per Person (USD)')
plt.title('Average Funding Gap vs Child Poverty Rate')
plt.legend()
plt.show()
```
